```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, lit
```

```python
spark = SparkSession.builder.appName("df_to_RDD").getOrCreate()
```

```python
df = spark.read.options(header = True, inferSchema = True).csv("StudentData.csv")
df.show()
```

```python
type(df)
```

```python
rdd = df.rdd
```

```python
type(rdd)
```

```python
rdd.collect()
```

```python
rdd.filter(lambda x: x['gender']=="Male").collect()
```

```python
rdd.filter(lambda x: x['age']==28 and x['gender'] =='Male').collect()
```
